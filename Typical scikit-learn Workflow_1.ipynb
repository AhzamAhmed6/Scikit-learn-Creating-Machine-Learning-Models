```python
import pandas as pd
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

### we will predict the "target" colum (last column) from the data of all other column

```python
# x = complete table except target coluumn
x=heart_disease.drop("target", axis=1) # axis=1 for rows.....axis=0 for column
x
```

```python
# y = "target" column
y= heart_disease["target"]
y
```

### selecting the right model from the sklearn library

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier() # by default it make 100 trees
clf.get_params() # to see the methord contained by clf
```

```python
# It split data x,y. Train it and test it
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2) # 0.2 out of 1.0 data is used for testing and 0.8 is used for training ml model
```

```python
# hey classification model RandomForest find the pattern in the train data
clf.fit(x_train, y_train); # fit the train data in to RandomForest classifier
```

```python
# now x has --> 303 rows × 13 columns <----- 100%
x
```

```python
# x_trian has --> 242 rows × 13 columns <----- 80%
x_train
```

```python
# x_test has ----> 61 rows × 13 columns <----- 20%
x_test
```

#### similar in the case of y, y_train, y_test

```python
# making preductions using inputs and reuired outputs
y_preds=clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# checkin the accuraccy of prediction of our model (1.0 = 100%)
clf.score(x_train, y_train)
```

```python
# test it's prediction on test data
clf.score(x_test, y_test)
```

so the model give his 100% in finding the pattern between input and required output, but on test data it found that it's 83.6%
results are correct based on it's predictions

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# there are 3 ways we can check our reult beside .score methord
```

```python
# compare the test and the prediction
print(classification_report(y_test, y_preds))  #----------> 1st methord
```

```python
print(confusion_matrix(y_test, y_preds))   #----------> 2nd methord
```

```python
print(accuracy_score(y_test, y_preds))  #----------> 3rd methord
```

# improving our accuracy

```python
import numpy as np
np.random.seed(0)
for i in range (10, 100, 10):
    clf=RandomForestClassifier(i).fit(x_train, y_train) # clf store best logic made by random forest
    print(f"The accuracy with {i} numbers of trees is {clf.score(x_test, y_test)*100}")
```
