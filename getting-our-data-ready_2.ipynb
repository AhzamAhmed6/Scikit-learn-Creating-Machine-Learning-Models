# now we take all the non_numeric data and convert it in to numeric data 

```python
from sklearn.preprocessing import OneHotEncoder #   to place 1 and 0

from sklearn.compose import ColumnTransformer # to convert columns in to rows

categorical_features=["Make", "Colour", "Doors"] # take doors in categorical features because of its grouping

one_hot=OneHotEncoder()  # making oject of OneHotEncoder class

transformer= ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")
# we wants to perform "one_hot" function on the "categorical features", and the remains will be "pass through"

transformed_x=transformer.fit_transform(x) # fit the x table in transformer variable
transformed_x

# Summry:
#     hey "ColumnTransformer" take the "OneHotEncoder" and apply it on the "categorical_features" and for the remaining features just pass them
```

```python
#put the above garbage table in to data frame
pd.DataFrame(transformed_x)
```
